# CAFA6 Submit Blend

Submission wrapper for GOA + ProtT5 blend output.

```python
import os
import shutil

INPUT_PATH = '/kaggle/input/cafa6-goa-prott5-blend/submission.tsv'
OUTPUT_PATH = 'submission.tsv'

print(f"Looking for: {INPUT_PATH}")

input_dir = '/kaggle/input/cafa6-goa-prott5-blend'
if os.path.exists(input_dir):
    print(f"\nContents of {input_dir}:")
    for f in os.listdir(input_dir):
        full_path = os.path.join(input_dir, f)
        if os.path.isfile(full_path):
            size = os.path.getsize(full_path)
            print(f"  {f}: {size:,} bytes")
        else:
            print(f"  {f}/ (directory)")
else:
    print(f"ERROR: {input_dir} not found!")
    print("\nAvailable inputs:")
    for d in os.listdir('/kaggle/input'):
        print(f"  /kaggle/input/{d}")
```

```python
if os.path.exists(INPUT_PATH):
    shutil.copy(INPUT_PATH, OUTPUT_PATH)
    size = os.path.getsize(OUTPUT_PATH)
    print(f"SUCCESS: Copied to {OUTPUT_PATH} ({size:,} bytes)")
    
    with open(OUTPUT_PATH, 'r') as f:
        lines = sum(1 for _ in f)
    print(f"Total predictions: {lines:,}")
else:
    print(f"ERROR: {INPUT_PATH} not found!")
```

```python
import pandas as pd

if os.path.exists(OUTPUT_PATH):
    df = pd.read_csv(OUTPUT_PATH, sep='\t', header=None, 
                     names=['protein_id', 'go_term', 'score'],
                     nrows=20)
    print("Sample predictions:")
    print(df)
    
    print(f"\nScore range: [{df['score'].min():.3f}, {df['score'].max():.3f}]")
```
