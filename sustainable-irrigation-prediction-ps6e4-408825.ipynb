```python
# TABLE STYLING (FIXED: WHITE TEXT + HIGH CONTRAST)
def style_table(df):
    styled_df = df.style.set_table_styles([
        {"selector": "th", "props": [
            ("color", "#ffffff"),
            ("background-color", "#9b1c31"),
            ("font-weight", "bold"),
            ("font-size", "14px"),
            ("padding", "10px"),
            ("border", "1px solid #ffccd5"),
            ("text-shadow", "0 0 6px #000000")
        ]},
        {"selector": "td", "props": [
            ("text-align", "center"),
            ("color", "#ffffff"),
            ("background-color", "#1f2937"),
            ("border", "1px solid #374151"),
            ("padding", "8px")
        ]},
        {"selector": "table", "props": [
            ("background-color", "#111827"),
            ("border-radius", "12px")
        ]}
    ]).hide(axis="index")
    return styled_df.to_html()


# RANDOM COLOR
def generate_random_color():
    return "#{:02x}{:02x}{:02x}".format(
        random.randint(120, 255),
        random.randint(80, 200),
        random.randint(120, 255)
    )


# STYLED HEADINGS (FIXED VISIBILITY)
def styled_heading(text, gradient, text_color='#ffffff', border_color='#22c55e', font_size='24px'):
    return f"""
    <div style="
        background: {gradient};
        background-size: 300% 300%;
        animation: gradientMove 6s ease infinite;
        text-align: center;
        color: {text_color};
        padding: 22px 40px;
        border-radius: 25px;
        font-family: 'Poppins', sans-serif;
        font-weight: 800;
        font-size: {font_size};
        letter-spacing: 2px;
        border: 3px solid {border_color};
        box-shadow: 0px 0px 25px rgba(0,0,0,0.6);
        margin: 25px 0;
    ">
        {text}
    </div>
    <style>
        @keyframes gradientMove {{
            0% {{ background-position: 0% 50%; }}
            50% {{ background-position: 100% 50%; }}
            100% {{ background-position: 0% 50%; }}
        }}
    </style>
    """


# SECTION HEADER (FIXED)
def section_header(title, gradient):
    return f"""
    <div style="
        background: {gradient};
        color: #ffffff;
        text-align: center;
        padding: 12px 25px;
        border-radius: 12px;
        font-family: 'Poppins', sans-serif;
        font-weight: 700;
        font-size: 20px;
        margin-top: 15px;
        box-shadow: 0px 4px 12px rgba(0,0,0,0.5);
    ">
        {title}
    </div>
    """


# MAIN FUNCTION (UNCHANGED LOGIC)
def print_dataset_analysis(dataset, dataset_name, n_top=5, palette_index=0):

    main_gradients = [
        "linear-gradient(90deg, #111827, #7c2d12, #1f2937)",
        "linear-gradient(90deg, #1f2937, #9b1c31, #111827)",
        "linear-gradient(90deg, #0f172a, #7c2d12, #111827)"
    ]

    sub_gradients = [
        "linear-gradient(90deg, #9b1c31, #1f2937)",
        "linear-gradient(90deg, #7c2d12, #111827)",
        "linear-gradient(90deg, #1f2937, #9b1c31)"
    ]

    heading_gradient = main_gradients[palette_index % len(main_gradients)]
    sub_gradient = sub_gradients[palette_index % len(sub_gradients)]

    display(HTML(styled_heading(f"{dataset_name} Overview", heading_gradient)))

    # SHAPE
    display(HTML(section_header("📏 Shape of the Dataset", sub_gradient)))
    display(HTML(f"""
    <div style='
        color:#ffffff;
        font-family:Poppins;
        font-size:16px;
        font-weight:600;
        text-align:center;
        background:#111827;
        border:1px solid #374151;
        padding:10px;
        border-radius:10px;
        margin:10px 0;
    '>
        {dataset.shape[0]} rows × {dataset.shape[1]} columns
    </div>
    """))

    # HEAD
    display(HTML(section_header("👀 First Rows", sub_gradient)))
    display(HTML(style_table(dataset.head(n_top))))

    # STATS
    display(HTML(section_header("📊 Summary Statistics", sub_gradient)))
    display(HTML(style_table(dataset.describe())))

    # NULLS
    display(HTML(section_header("🚨 Null Values", sub_gradient)))
    null_counts = dataset.isnull().sum()
    if null_counts.sum() == 0:
        display(HTML("<p style='color:#22c55e;font-weight:600;'>✅ No null values found.</p>"))
    else:
        null_df = null_counts[null_counts > 0].to_frame("Null Values")
        null_df["Column"] = null_df.index
        display(HTML(style_table(null_df)))

    # DUPLICATES
    display(HTML(section_header("🔁 Duplicate Rows", sub_gradient)))
    display(HTML(f"<p style='color:#f87171;font-weight:600;'>{dataset.duplicated().sum()} duplicates found.</p>"))

    # DTYPE
    display(HTML(section_header("🔢 Data Types", sub_gradient)))
    dtypes_df = pd.DataFrame({
        "Column": dataset.columns,
        "Data Type": [dataset[col].dtype for col in dataset.columns]
    })
    display(HTML(style_table(dtypes_df)))

    # COLUMNS
    display(HTML(section_header("📋 Column Names", sub_gradient)))
    cols_html = ", ".join([
        f"<span style='color:#ffffff;font-weight:600'>{c}</span>"
        for c in dataset.columns
    ])
    display(HTML(f"<div style='color:white'>{cols_html}</div>"))

    # UNIQUE
    display(HTML(section_header("🔍 Unique Values", sub_gradient)))
    unique_df = pd.DataFrame({
        "Column": dataset.columns,
        "Unique Values": [dataset[col].nunique() for col in dataset.columns]
    })
    display(HTML(style_table(unique_df)))

    # INFO
    display(HTML(section_header("ℹ️ Dataset Info", sub_gradient)))
    buffer = io.StringIO()
    dataset.info(buf=buffer)
    display(HTML(f"<pre style='background:#111827;color:#ffffff;padding:12px;border-radius:10px'>{buffer.getvalue()}</pre>"))

# LOAD DATA
df_train = pd.read_csv("/kaggle/input/competitions/playground-series-s6e4/train.csv")
df_test = pd.read_csv("/kaggle/input/competitions/playground-series-s6e4/test.csv")
sample_sub = pd.read_csv("/kaggle/input/competitions/playground-series-s6e4/sample_submission.csv")
df_org = pd.read_csv("/kaggle/input/datasets/miadul/irrigation-water-requirement-prediction-dataset/irrigation_prediction.csv")  # if exists

# DISPLAY ALL
print_dataset_analysis(df_train, "Training Data", palette_index=0)
print_dataset_analysis(df_test, "Testing Data", palette_index=1)
print_dataset_analysis(sample_sub, "Sample Submission Data", palette_index=2)
print_dataset_analysis(df_org, "Original Dataset", palette_index=3)
```

# <p style="background-color:#7c2d12;font-family:'Georgia',serif;font-size:90%;text-align:center;color:#fde68a;border-radius:1000px 50px;border:4px solid #22c55e;padding:20px 40px;margin:15px 0;">Lets do Canyon Analysis by Comparing Real vs Synthetic Rainfall Distributions</p>

```python
def plot_canyon_analysis(train_df, org_df):
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np
    
    plt.style.use("dark_background")
    fig, axes = plt.subplots(2, 1, figsize=(18, 12), facecolor="#0f172a")
    
    # KDE DISTRIBUTION PLOT
    sns.kdeplot(
        org_df['Rainfall_mm'],
        ax=axes[0],
        label='Original Data',
        color='#38bdf8',
        linewidth=3,
        fill=True,
        alpha=0.35
    )
    sns.kdeplot(
        train_df['Rainfall_mm'],
        ax=axes[0],
        label='Synthetic Data',
        color='#f97316',
        linewidth=3,
        fill=True,
        alpha=0.35
    )
    axes[0].set_title(
        'Rainfall Distribution Comparison',
        fontsize=16,
        fontweight='bold',
        color='white'
    )
    axes[0].set_xlabel("Rainfall (mm)", fontsize=13, color="white")
    axes[0].set_ylabel("Density", fontsize=13, color="white")
    axes[0].legend(facecolor="#111827", edgecolor="white")
    axes[0].grid(True, alpha=0.25, linestyle='--')
    axes[0].set_facecolor("#111827")
    
    # IMPROVED SECOND PLOT WITH TWIN Y-AXIS
    bins = np.linspace(0, 2500, 80)
    hist_org, _ = np.histogram(org_df['Rainfall_mm'], bins=bins, density=True)
    hist_train, _ = np.histogram(train_df['Rainfall_mm'], bins=bins, density=True)
    ratio = hist_train / (hist_org + 1e-8)
    
    ax1 = axes[1]                   
    ax2 = ax1.twinx()                
    
    # Original Density 
    ax1.plot(
        bins[:-1],
        hist_org,
        color='#38bdf8',
        linewidth=2.8,
        label='Original Density',
        alpha=0.95
    )
    # Synthetic Density 
    ax1.plot(
        bins[:-1],
        hist_train,
        color='#f97316',
        linewidth=2.8,
        label='Synthetic Density',
        alpha=0.95
    )
    # Density Ratio 
    ax2.plot(
        bins[:-1],
        ratio,
        color='#ef4444',
        linewidth=3.2,
        label='Density Ratio (Synthetic / Original)'
    )
    
    ax1.axhline(0, color='white', linestyle='--', linewidth=1, alpha=0.3)
    ax2.axhline(1.0, color='white', linestyle='--', linewidth=1.5, alpha=0.7)
    
    # Styling
    ax1.set_title(
        'Density Comparison & Ratio (Synthetic vs Original)',
        fontsize=16,
        fontweight='bold',
        color='white'
    )
    ax1.set_xlabel("Rainfall (mm)", fontsize=13, color="white")
    ax1.set_ylabel("Density", fontsize=13, color="#38bdf8")         
    ax2.set_ylabel("Density Ratio", fontsize=13, color="#ef4444")   
    
    # Legends
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, 
               facecolor="#111827", edgecolor="white", loc='upper right')
    
    ax1.grid(True, alpha=0.25, linestyle='--')
    ax1.set_facecolor("#111827")
    
    plt.suptitle(
        'Synthetic Generator Sampling Deficit Analysis',
        fontsize=20,
        fontweight='bold',
        color='white'
    )
    plt.tight_layout()
    plt.show()

# Call the function
plot_canyon_analysis(df_train, df_org)
```

# <p style="background-color:#7c2d12;font-family:'Georgia',serif;font-size:90%;text-align:center;color:#fde68a;border-radius:1000px 50px;border:4px solid #22c55e;padding:20px 40px;margin:15px 0;">Lets Visualize Key Numeric Features Distribution by Irrigation Need</p>

```python
def plot_numeric_distribution(df_train, target_col):
    import matplotlib.pyplot as plt
    import seaborn as sns

    key_numerics = [
        'Soil_Moisture',
        'Rainfall_mm',
        'Temperature_C',
        'Humidity'
    ]

    plt.style.use("dark_background")

    fig, axes = plt.subplots(2, 2, figsize=(16, 10), facecolor="#0f172a")
    axes = axes.ravel()

    palette = ['#38bdf8', '#f97316', '#22c55e']  # low, medium, high

    for i, col in enumerate(key_numerics):
        ax = axes[i]

        sns.boxplot(
            data=df_train,
            x=target_col,
            y=col,
            ax=ax,
            palette=palette,
            width=0.6,
            linewidth=1.5
        )
        ax.set_title(
            f'{col} by {target_col}',
            fontweight='bold',
            color='white',
            pad=12
        )

        ax.set_xlabel(target_col, color="white")
        ax.set_ylabel(col, color="white")
        ax.set_facecolor("#111827")
        ax.grid(True, alpha=0.25)
        ax.tick_params(colors='white')
        
        # MEAN & MAX VALUES
        grouped = df_train.groupby(target_col)[col]

        means = grouped.mean()
        maxs = grouped.max()

        for j, cls in enumerate(means.index):
            mean_val = means[cls]
            max_val = maxs[cls]

            # MEAN label
            ax.text(
                j - 0.12,
                mean_val,
                f"Mean: {mean_val:.2f}",
                ha='center',
                va='bottom',
                fontsize=9,
                color='#38bdf8',
                fontweight='bold',
                bbox=dict(facecolor='#111827', edgecolor='#38bdf8', boxstyle='round,pad=0.25', alpha=0.8)
            )

            # MAX label
            ax.text(
                j + 0.12,
                max_val,
                f"Max: {max_val:.2f}",
                ha='center',
                va='bottom',
                fontsize=9,
                color='#f97316',
                fontweight='bold',
                bbox=dict(facecolor='#111827', edgecolor='#f97316', boxstyle='round,pad=0.25', alpha=0.8)
            )

    fig.suptitle(
        'Key Numeric Feature Distribution (Mean & Max Analysis)',
        fontsize=18,
        fontweight='bold',
        color='white',
        y=0.98
    )

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()


plot_numeric_distribution(df_train, target_col)
```

# <p style="background-color:#7c2d12;font-family:'Georgia',serif;font-size:90%;text-align:center;color:#fde68a;border-radius:1000px 50px;border:4px solid #22c55e;padding:20px 40px;margin:15px 0;">Mulching Shield Effect analysis | Impact on High Irrigation Need</p>

```python
def analyze_mulching_shield(train_df, org_df, target_col):
    import pandas as pd
    from IPython.display import display, HTML

    def mulch_stats(df, label):
        rates = df.groupby('Mulching_Used')[target_col].apply(
            lambda x: (x == 'High').mean() * 100
        )

        multiplier = rates['No'] / rates['Yes']

        return {
            'Dataset': label,
            'High_NoMulch (%)': rates['No'],
            'High_WithMulch (%)': rates['Yes'],
            'Protection_Multiplier': multiplier
        }

    # compute results
    df_mulch = pd.DataFrame([
        mulch_stats(org_df, 'Original'),
        mulch_stats(train_df, 'Train')
    ])

    df_mulch = df_mulch.round(2)
    styled = df_mulch.style.set_table_styles([
        {"selector": "th", "props": [
            ("background-color", "#111827"),
            ("color", "#38bdf8"),
            ("font-weight", "bold"),
            ("text-align", "center"),
            ("padding", "10px"),
            ("border", "1px solid #38bdf8")
        ]},
        {"selector": "td", "props": [
            ("background-color", "#0f172a"),
            ("color", "white"),
            ("text-align", "center"),
            ("padding", "8px"),
            ("border", "1px solid #1f2937")
        ]}
    ])

    display(styled)
analyze_mulching_shield(df_train, df_org, target_col)
```

">
<h3 style="color:#22c55e;">Overall Conclusion</h3>

<ul>
<li>The Train dataset is not identical to the Original:</li>
<li>It slightly increases High without mulch</li>
<li>It decreases High with mulch</li>

# <p style="background-color:#7c2d12;font-family:'Georgia',serif;font-size:90%;text-align:center;color:#fde68a;border-radius:1000px 50px;border:4px solid #22c55e;padding:20px 40px;margin:15px 0;">Feature Engineering</p>

```python
def apply_enhanced_feature_engineering(df):
    d = df.copy()
    
    # Original strong features
    high_score = (d["Soil_Moisture"] < 25).astype(int) * 2 + (d["Rainfall_mm"] < 300).astype(int) * 2 + \
                 (d["Temperature_C"] > 30).astype(int) + (d["Wind_Speed_kmh"] > 10).astype(int)
    low_score = (d["Crop_Growth_Stage"] == "Harvest").astype(int) * 2 + (d["Crop_Growth_Stage"] == "Sowing").astype(int) * 2 + \
                (d["Mulching_Used"] == "Yes").astype(int)
    d["physical_score"] = high_score - low_score
    
    # Adaptive rounding
    num_cols = ["Soil_pH", "Soil_Moisture", "Organic_Carbon", "Electrical_Conductivity",
                "Temperature_C", "Humidity", "Rainfall_mm", "Sunlight_Hours",
                "Wind_Speed_kmh", "Field_Area_hectare", "Previous_Irrigation_mm"]
    for col in num_cols:
        max_val = d[col].max()
        decimals = 3 if max_val < 10 else 2 if max_val < 100 else 1
        d[f"{col}_rounded"] = d[col].round(decimals)
    
    # Forensic flags
    d['canyon_low_rain'] = (d['Rainfall_mm'] < 463).astype(int)
    d['canyon_high_rain'] = ((d['Rainfall_mm'] >= 1902) & (d['Rainfall_mm'] <= 2007)).astype(int)
    d['q3_rain_breach_flag'] = ((d['Soil_Moisture'] < 25) & (d['Rainfall_mm'] >= 300)).astype(int)
    d['mulch_shield_interaction'] = (d["Mulching_Used"] == "No").astype(int) * d["Soil_Moisture"]
    d['temp_humidity_ratio'] = d['Temperature_C'] * (1 - d['Humidity']/100)          
    d['rain_moisture_ratio'] = d['Rainfall_mm'] / (d['Soil_Moisture'] + 1)
    d['temp_decimal'] = (d['Temperature_C'] * 10 % 10).astype(int)
    d['humidity_decimal'] = (d['Humidity'] * 10 % 10).astype(int)
    d['rainfall_bin'] = pd.cut(d['Rainfall_mm'], bins=[0, 463, 954, 1467, 2054, 2500], labels=[0,1,2,3,4]).astype(int)
    for g in ['Crop_Type', 'Region']:
        d[f'{g}_mean_rain'] = d.groupby(g)['Rainfall_mm'].transform('mean')
        d[f'{g}_mean_moisture'] = d.groupby(g)['Soil_Moisture'].transform('mean')
    
    return d
```

# <p style="background-color:#7c2d12;font-family:'Georgia',serif;font-size:90%;text-align:center;color:#fde68a;border-radius:1000px 50px;border:4px solid #22c55e;padding:20px 40px;margin:15px 0;">LightGBM Model Training with In-Fold Target Encoding & Bias Optimization</p>

```python
print("\n" + "="*80)
print("BUILDING FURTHER IMPROVED LIGHTGBM PIPELINE")
print("="*80)

X_raw = df_train.drop(['id', target_col, 'target_numeric'], axis=1, errors='ignore')
y_numeric = df_train['target_numeric'].values.astype(int)
test_ids = df_test['id'].copy()

cat_features = ["Soil_Type", "Crop_Type", "Crop_Growth_Stage", "Season", "Irrigation_Type", "Water_Source", "Mulching_Used", "Region"]
num_features = ["Soil_pH", "Soil_Moisture", "Organic_Carbon", "Electrical_Conductivity", "Temperature_C", "Humidity", "Rainfall_mm", "Sunlight_Hours", "Wind_Speed_kmh", "Field_Area_hectare", "Previous_Irrigation_mm"]

print("Applying Enhanced Feature Engineering...")
X_full = apply_enhanced_feature_engineering(X_raw)
X_test_full = apply_enhanced_feature_engineering(df_test.drop('id', axis=1))

N_FOLDS = 5
cv = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=42)
oof_predictions = np.zeros((len(X_full), 3))
test_predictions = np.zeros((len(X_test_full), 3))

lgb_hyperparams = {
    'objective': 'multiclass', 'num_class': 3, 'metric': 'multi_logloss',
    'learning_rate': 0.038, 'num_leaves': 255, 'min_child_samples': 35,
    'subsample': 0.88, 'colsample_bytree': 0.68, 'n_estimators': 2600,
    'reg_alpha': 0.08, 'reg_lambda': 0.12,
    'random_state': 42, 'verbose': -1, 'n_jobs': -1
}

print("Training with In-Fold Target Encoding")
for fold_idx, (train_idx, val_idx) in enumerate(cv.split(X_full, y_numeric)):
    X_tr = X_full.iloc[train_idx].copy()
    y_tr = y_numeric[train_idx]
    X_val = X_full.iloc[val_idx].copy()
    y_val = y_numeric[val_idx]
    X_te = X_test_full.copy()
    
    te_cols = cat_features + [c for c in X_full.columns if '_rounded' in c or '_bin' in c or '_decimal' in c]
    encoder = TargetEncoder(target_type="multiclass", random_state=42)
    
    enc_train = pd.DataFrame(encoder.fit_transform(X_tr[te_cols], y_tr).astype(np.float32), index=X_tr.index)
    enc_val = pd.DataFrame(encoder.transform(X_val[te_cols]).astype(np.float32), index=X_val.index)
    enc_test = pd.DataFrame(encoder.transform(X_te[te_cols]).astype(np.float32), index=X_te.index)
    
    enc_cols = [f"target_enc_{i}" for i in range(enc_train.shape[1])]
    enc_train.columns = enc_cols; enc_val.columns = enc_cols; enc_test.columns = enc_cols
    
    base_cols = [c for c in X_full.columns if c not in te_cols]
    X_train_final = pd.concat([X_tr[base_cols], enc_train], axis=1)
    X_val_final = pd.concat([X_val[base_cols], enc_val], axis=1)
    X_test_final = pd.concat([X_te[base_cols], enc_test], axis=1)
    
    sample_weights = compute_sample_weight("balanced", y_tr)
    
    model = lgb.LGBMClassifier(**lgb_hyperparams)
    model.fit(X_train_final, y_tr, sample_weight=sample_weights,
              eval_set=[(X_val_final, y_val)],
              callbacks=[lgb.early_stopping(60, verbose=False)])
    
    oof_predictions[val_idx] = model.predict_proba(X_val_final)
    test_predictions += model.predict_proba(X_test_final) / N_FOLDS
    
    print(f"  Fold {fold_idx+1} BA: {balanced_accuracy_score(y_val, np.argmax(oof_predictions[val_idx], axis=1)):.5f}")

raw_oof = balanced_accuracy_score(y_numeric, np.argmax(oof_predictions, axis=1))
print(f"\nRaw OOF Balanced Accuracy: {raw_oof:.6f}")

# Bias Tuning
print("\nAdvanced Bias Tuning...")
def apply_bias(proba, bias):
    return np.argmax(np.log(np.clip(proba, 1e-15, 1.0)) + bias, axis=1)

def bias_objective(trial):
    bias_vec = np.array([trial.suggest_float(f"b{i}", -2.5, 2.5) for i in range(3)])
    return balanced_accuracy_score(y_numeric, apply_bias(oof_predictions, bias_vec))

study = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler(seed=42))
study.optimize(bias_objective, n_trials=120, show_progress_bar=False)

best_bias = np.array([study.best_params[f"b{i}"] for i in range(3)])
tuned_oof = study.best_value
print(f"Tuned OOF Balanced Accuracy: {tuned_oof:.6f} (+{tuned_oof-raw_oof:.5f} lift)")

final_preds = apply_bias(test_predictions, best_bias)
submission = pd.DataFrame({'id': test_ids, target_col: [inverse_mapping[p] for p in final_preds]})
submission.to_csv('submission.csv', index=False)
print("submission.csv created sucessfully!")
```

# <p style="background-color:#7c2d12;font-family:'Georgia',serif;font-size:90%;text-align:center;color:#fde68a;border-radius:1000px 50px;border:4px solid #22c55e;padding:20px 40px;margin:15px 0;">Lets Visualize Feature Importance</p>

```python
def save_and_plot_importance(oof_predictions, test_predictions, best_bias, X_train_final, model):
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    def calibrate_probs(probs, bias):
        log_p = np.log(np.clip(probs, 1e-15, 1.0)) + bias
        exp_p = np.exp(log_p)
        return exp_p / exp_p.sum(axis=1, keepdims=True)

    print("\nSaving calibrated probabilities...")

    np.save(
        'oof_probs_final.npy',
        calibrate_probs(oof_predictions, best_bias)
    )

    np.save(
        'test_probs_final.npy',
        calibrate_probs(test_predictions, best_bias)
    )

    print("Ensemble files saved successfully.")

    # FEATURE IMPORTANCE
    imp_df = pd.DataFrame({
        'feature': X_train_final.columns,
        'importance': model.feature_importances_
    })

    imp_df = imp_df.sort_values('importance', ascending=False).head(20)
    plt.style.use("dark_background")

    plt.figure(figsize=(12, 8), facecolor="#0f172a")

    sns.barplot(
        data=imp_df,
        x='importance',
        y='feature',
        palette=['#38bdf8', '#f97316', '#ef4444']
    )

    plt.title(
        'Top 20 Important Features',
        fontsize=18,
        fontweight='bold',
        color='white',
        pad=15
    )

    plt.xlabel("Importance", color="white")
    plt.ylabel("Feature", color="white")

    plt.gca().set_facecolor("#111827")
    plt.gca().tick_params(colors='white')

    plt.grid(True, alpha=0.2)

    plt.tight_layout()
    plt.show()
save_and_plot_importance(
    oof_predictions,
    test_predictions,
    best_bias,
    X_train_final,
    model
)
```
