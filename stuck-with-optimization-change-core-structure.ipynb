# Beyond Bbox Optimization: Building New Architectures for Tree Packing

## The Problem with Incremental Optimization

It is strange that so many people spend 100% of their time just running bbox binaries and trying to improve the score in 5-8 digits after the decimal point.

**This is a way to nowhere.**

Bbox is great to **improve** the architecture, but not to **build** a new one.

If you want significant changes - try to create a new base architecture and use bbox to refine it. You will be surprised by the results.

## What is an Alternative?

Just try another basic architecture. For example - a mix of two SA (Simulated Annealing) runs with a **2-trees unicell core**, each one with independent parameters (pair, a, b, etc.), but same target functuion which minimize joint square side.

It is not so difficult to implement. Just take existing public code for SA and update it to support two independent blocks.

You can get a new solution in 5 minutes of execution on your laptop and it will be better than weeks of running optimizations based on existing core structure. For example, if you need to get 130 trees, use a primary block with 5 columns and 12 rows + a secondary block with 1 column and 5 rows:



It is significantly better, and many small solutions could be improved by just running https://www.kaggle.com/code/chistyakov/santa-2025-simple-optimization-new-slow-version on top of it.

---

## Common Pitfalls to Avoid

###  Pitfall 1: Optimizing a Non-Optimal Architecture

Don't start optimizing until you understand the architecture. You'll waste time on the wrong parameters.

###  Pitfall 2: Using Public Solutions As-Is Without Trying to Change Them

Public solutions are good to start, but don't run them many times without significant changes.

---

## Conclusion

**Stop optimizing. Start architecting.**

The real gains come from new structures, not from squeezing the last 0.000001 from an existing one. Use bbox as an improvement tool to refine the solution, not as your primary strategy.

Build something new. Validate it. Refine it. Then move to the next architecture.
