```python
# Setup

def f_convert(OPTION):
    OP    = OPTION['N']
    LB    = [e['Lb']        for e in OPTION['SOLUTIONs']]
    ES    = [e['Solution']  for e in OPTION['SOLUTIONs']]
    WTS   = [e['weight']    for e in OPTION['SOLUTIONs']]
    s_wts = OPTION['sub-weights'] if "sub-weights"   in OPTION else []
    SORT  = OPTION['Sort'] if 'Sort' in OPTION else ""
    FS    = ['subm_'  + e.split('_')[1] + '.csv' for e in ES]
    return OP, LB, ES, FS, WTS, s_wts, SORT


import matplotlib.pyplot as plt
import pandas as pd


def select_f_ensemble(OPTION):

    N_OPTION, LB, ENSEMBLE_SOLUTIONS, FILES_SUBM, WEIGHTS, sub_WEIGHTS, SORT = f_convert(OPTION)

    __target = 'Calories'
    
    def visual_distribution_target(df, target):
        plt.figure(figsize=(5, 2))
        cmap = ["Silver","Navy","Black","Crimson","#777abc"]
        df[target].plot(color=cmap[0], linewidth=0.1, title=f"{target} Distribution")
        plt.axhline(df[target].mean(), c="w", ls="-", lw=2.5)
        plt.show()
    
    def EoS_N(lbs        = LB,
              solution   = ENSEMBLE_SOLUTIONS,
              weights    = WEIGHTS,
              files_subm = FILES_SUBM,
              option     = OPTION):
        def f_ensemble(df,cols,wts):
            return sum([df[cols[i]]*wts[i] for i in range(len(wts))])
        suffix = 's'
        soluts = [solut.replace("SOLUTION_","").replace(suffix,'') for solut in solution]
        target = 'Calories'
        list_df = [pd.read_csv(files_subm[i]) for i in range(len(FILES_SUBM))]
        targets = target.replace('Calories',"m")
        for i in range(len(FILES_SUBM)): 
            list_df[i] = list_df[i].rename(columns={target:f'{soluts[i]}'})                    
        dfs = pd.merge(list_df[0], list_df[1], on=['id'])
        for i in range(2,len(FILES_SUBM)): 
            dfs = pd.merge(dfs,list_df[i],on=['id'])
        cols = [f'{soluts[i]}' for i in range(len(FILES_SUBM))]
        dfs[target] = dfs.apply(lambda x:f_ensemble(x,cols,weights), axis=1)
        #print(f'\nLB: {lbs}  \n\nweights: {weights}\n')
        dfs = dfs.rename(columns = {'Calories': "Ensemble"})
        dfs.insert(loc=1, column=' _ ', value=['   ']*250_000)
        dfs[' _ '] = dfs[' _ '].astype(str)
        nr = 51; pd.set_option('display.max_rows', nr);
        pd.set_option('display.float_format', '{:.2f}'.format)
        display(dfs.head(nr))
        pd.set_option('display.float_format', '{:.7f}'.format)
        dfs = dfs.rename(columns = {"Ensemble":'Calories'})
        return dfs


    def EoS_MD(lbs        = LB,
               solution   = ENSEMBLE_SOLUTIONS,
               weights    = WEIGHTS,
               files_subm = FILES_SUBM,
               option     = OPTION):

        from statistics import median

        def f_ensembleMD(df,cols):
            return median([df[col] for col in cols])
         
        suffix = 's'
        soluts = [solut.replace("SOLUTION_","").replace(suffix,'') for solut in solution]
        target = 'Calories'
        list_df = [pd.read_csv(files_subm[i]) for i in range(len(FILES_SUBM))]
        targets = target.replace('Calories',"m")
        for i in range(len(FILES_SUBM)): 
            list_df[i] = list_df[i].rename(columns={target:f'{soluts[i]}'})                    
        dfs = pd.merge(list_df[0], list_df[1], on=['id'])
        for i in range(2,len(FILES_SUBM)): 
            dfs = pd.merge(dfs,list_df[i],on=['id'])
        cols = [f'{soluts[i]}' for i in range(len(FILES_SUBM))]
        dfs[target] = dfs.apply(lambda x:f_ensembleMD(x,cols), axis=1)
        #print(f'\nLB: {lbs}  \n\nweights: {weights}\n')
        dfs = dfs.rename(columns = {'Calories': "Ensemble"})
        dfs.insert(loc=1, column=' _ ', value=['   ']*250_000)
        dfs[' _ '] = dfs[' _ '].astype(str)
        nr = 11; pd.set_option('display.max_rows', nr);
        pd.set_option('display.float_format', '{:.2f}'.format)
        display(dfs.head(nr))
        pd.set_option('display.float_format', '{:.7f}'.format)
        dfs = dfs.rename(columns = {"Ensemble":'Calories'})
        return dfs
    
    
    def EoS_DA(lbs         = LB,
               solution    = ENSEMBLE_SOLUTIONS,
               weights     = WEIGHTS,
               sub_weights = sub_WEIGHTS,
               sort        = SORT,
               files_subm  = FILES_SUBM,
               option      = OPTION):
        def read_subm(path): 
            return pd.read_csv(path).rename(
                columns={__target: path.replace('.csv',''),'target': path.replace('.csv','')})                                         
        df_subm = [read_subm(files_subm[i]) for i in range(len(files_subm))]
        df_subm_s = pd.merge(df_subm[0], df_subm[1], on=['id'])
        for i in range(2,len(files_subm)): df_subm_s = pd.merge(df_subm_s,df_subm[i],on=['id'])
        cols = [col for col in df_subm_s.columns if col != "id"]
        prefix,target = "subm_",__target
        short_name_cols = [c.replace(prefix,'') for c in cols]
        correct_scales  = [w for w in sub_WEIGHTS]
        def alls(x, cs=cols, p=prefix,sort=sort):
            tes={c:x[c] for c in cs}.items()
            all_soluts_sorted = [
                t[0].replace(p,'') 
                for t in sorted(tes, key=lambda _:_[1],reverse=True if sort=='descending' else False)]
            return all_soluts_sorted
        def two_weights(x,p=prefix,cs=cols,W=weights,sncs=short_name_cols,cw=correct_scales):
            ic = [x['alls'].index(c) for c in sncs]
            cS  = [x[cols[ii]]*(W[ii]+cw[ic[ii]]) for ii in range(len(cols))]
            return sum(cS)
        df_subm_s['alls']   = df_subm_s.apply(lambda x: alls(x),          axis=1)
        df_subm_s[__target] = df_subm_s.apply(lambda x: two_weights(x),   axis=1)
        soluts = [solut.replace("SOLUTION_","") for solut in solution]
        #print(f'\nLB: {lbs}  \n\nweights: {weights} \n\nsub-weights: {sub_weights}\n')
        schema_rename = {
            old_name_cols : new_short_name_cols 
            for old_name_cols, new_short_name_cols in zip(cols, short_name_cols)}
        dfs = df_subm_s
        dfs = dfs.rename(columns = schema_rename)
        dfs = dfs.rename(columns = {'Calories': "Ensemble"})
        dfs.insert(loc=1, column=' _ ', value=['   ']*250_000)
        dfs[' _ '] = dfs[' _ '].astype(str)
        nr = 11; pd.set_option('display.max_rows', nr);
        pd.set_option('display.float_format', '{:.2f}'.format)

        vcols = ['id'] + ['alls'] + ['Ensemble']

        vcols = ['id'] + [' _ '] + short_name_cols + [' _ '] + ['alls'] + [' _ '] + ['Ensemble']
        
        dfs = dfs[vcols]
        display(dfs.head(nr))
        pd.set_option('display.float_format', '{:.7f}'.format)
        dfs = dfs.rename(columns = {"Ensemble":'Calories'})
        return dfs


    if OPTION['sub-weights']=="median":
        f_ensemble = EoS_MD
        print('\n', f_ensemble, f' median..')
    
    else:
        f_ensemble = EoS_DA if "Sort" in OPTION else EoS_N
        print('\n', f_ensemble, f' sort={SORT}')

    return OPTION, f_ensemble


def ensemble_DA(OPTION, N, type_sort):
    option = OPTION
    option['Sort'] = type_sort
    OPTION, f_ensemble = select_f_ensemble(option)
    dfs  =  f_ensemble()
    subm =  dfs [['id','Calories']]
    subm.to_csv(f'Datt_{N}.csv', index=False) 
    return subm
```
