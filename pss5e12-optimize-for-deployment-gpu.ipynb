```python
for i in ['family_history_diabetes','hypertension_history','cardiovascular_history']:
    train[i]=train[i].replace({0:'No',
                              1:'Yes'})
    test[i]=test[i].replace({0:'No',
                              1:'Yes'})
train.info()
```

```python
counter=0
for i in test.select_dtypes(include=['object']).columns.tolist():
    if (len(list(set(train[i].unique().tolist())^set(test[i].unique().tolist())))!=0):
        print(i ,'need to be worked on')
        counter+=1
    else:
        continue
if(counter==0):
    print('No work needed')
```

```python
label = 'diagnosed_diabetes'
train[label].value_counts()
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
predictor = TabularPredictor(label=label,eval_metric ='roc_auc',
                            problem_type="binary").fit(train,presets='optimize_for_deployment',
                                                                    time_limit=3600*9,verbosity=3,
                                                       ag_args_fit={'num_gpus': 1}
                                                      )
results = predictor.fit_summary()
```

```python
predictor.leaderboard()
```

```python
df = predictor.predict_proba(test)
df.head()
```

```python
sol=pd.read_csv('/kaggle/input/playground-series-s5e12/sample_submission.csv')
sol.head()
```

```python
sol[label]=df[1]
sol.head()
```
